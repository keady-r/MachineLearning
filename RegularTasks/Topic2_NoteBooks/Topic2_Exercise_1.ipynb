<b> what is an Absolute Value?
 
In short, an absolute value is the distance a number is from zero on a line without considering the direction of the value i.e. whether the number is positive or negative. In the example below the absolute value is 5 in both cases as 5 is the number of positions between -5 and zero and +5 and zero. [2]

# 2. Calculations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([-1.2, 1.2])
np.absolute(x)
np.absolute(1.2 + 1j)
x = np.linspace(-10, 10, 101)
plt.plot(x, np.absolute(x))
plt.show()
```
